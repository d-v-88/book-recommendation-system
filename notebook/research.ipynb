So for class studying algorithm and collaborative filtering we need to convert this data into pivot table 

```python
#lets create a pivot table
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values='rating')
```

So pandas has its own pivot table. Here i used 3 values in pivot table user id, rating of the book and book title to get the relation and collaborative filtering between each user and their ratings

so before training the model i need to convert the pivot table to csr matrix. it halps in computation.

so i will use k nearest neighbour and it accepts csr matrix

so to do this i will import scipy it has sparse and csr matrix

```python
from scipy.sparse import csr_matrix
```

insid this i passed the pivot table

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
type(book_sparse)
```

now the csr martix is done. i will import the nearest neighbour. so in this k-nearest neig.... we have different algos i will use brute

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

# testing model


```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)

    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            if j == book_name:
                print(f"You searched '{book_name}' \n")
                print("The suggestion books are: \n")
            else: 
                print(j)
```
