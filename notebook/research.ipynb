A user can read multipple books. Now we will check that 

```python
ratings['user_id'].value_counts()
```

So here we got to types of users the top users rates a lots of books so first user has rated 13602 books. and below are the users who has rated 1 book. So for the analysis in this project i will use people who have atleast rated 200 plus books. the people who rated books below 200 are the outliers. This is called Feature Engineering. We can also take more than 300, 400 but at 200 i was getting more accuracy.

```python
# Lets store users who has at least rated more than 200 books
x = ratings['user_id'].value_counts() > 200
```

```python
x
```

```python
x[x].shape
```

So from so many users we extracted users who have rated more books and those are 899 users

Now i will get the index number

```python
y = x[x].index
```

```python
y
```

Now i will push only the users who have rated 200 plus times in the ratings variable

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head()
```

```python
# Now join ratings with books
ratings_with_books = ratings.merge(books, on='ISBN')
```

```python
ratings_with_books.shape
```

```python
ratings_with_books.head()
```

Now i want which books have got how much rating

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
# Lets rename the column name
number_rating.rename(columns={'rating': 'num_of_rating'}, inplace=True)
```

```python
number_rating.head()
```

Now i will merge this number of ratings in the entire data column

```python
final_rating = ratings_with_books.merge(number_rating, on='title')
```

```python
final_rating.head()
```

So here we are manipulating the data with features. So this is called feature engineering.
